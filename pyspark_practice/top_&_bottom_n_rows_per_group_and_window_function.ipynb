```python
import pyspark, os, sys
from pyspark.sql import *
from pyspark import SparkConf,SparkContext
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import types
from pyspark import StorageLevel
from pyspark.sql.window import Window
from pyspark.sql.functions import col, row_number

spark=SparkSession.builder.getOrCreate()

data=[('mike','maths',80,'P',90),
      ('mike','python',75,'P',80),
      ('mike','science',40,'F',60),
      ('nancy','maths',56,'P',70),
      ('nancy','python',32,'F',63),
      ('nancy','science',90,'P',85),
      ('prince','maths',74,'P',68),
      ('prince','python',88,'P',82),
      ('prince','science',77,'P',89),
     ]

schema=StructType([
        StructField('name',StringType(), True),
        StructField('subject', StringType(), True),
        StructField('marks', IntegerType(), True),
        StructField('status', StringType(), True),
        StructField('attendance', IntegerType(), True)
])

df=spark.createDataFrame(data=data,schema=schema)
#df.show()

#Here we are partitioning the records based on 'name' column and ordering the records based on 'marks' column in descending order 
#and assigning row_number() to each record within the partition
window_df=Window.partitionBy("name").orderBy(col("marks").desc())
df_new=df.withColumn("row", row_number().over(window_df)).orderBy("name","row")
df_new.show()

#Here we are getting top 1 rows from each partition
df_filter=df_new.filter(col('row')<=1)
df_filter.show()

#create rank within each group of subject
window_df2=Window.partitionBy("subject").orderBy(col("marks").desc())
df_new2=df.withColumn("rank",row_number().over(window_df2)).orderBy("rank", "name")
display(df_new2)

#to get top N rows from above dataframe
df_filter2=df_new2.filter(col("rank")<=1)
df_filter.show()


#reverse tha above dataframe to get bottom N rows per group
window_df3=Window.partitionBy("subject").orderBy(col("marks"))
df_new3=df.withColumn("reverse_rank", row_number().over(window_df3)).orderBy("name", "reverse_rank")
df_new3.show()

#to get top 1 row from above dataframe
df_filter3=df_new3.filter(col("reverse_rank")<=1)
df_filter3.show()
```

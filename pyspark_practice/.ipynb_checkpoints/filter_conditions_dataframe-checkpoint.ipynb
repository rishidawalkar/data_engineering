```python
import pyspark, os, sys
```

```python
from pyspark.sql import *
```

```python
from pyspark import SparkConf,SparkContext
```

```python
spark=SparkSession.builder.getOrCreate()
```

```python
df=spark.read.format("csv").option("header",True).load("sample1.csv")
```

```python
df.show()
```

```python
df.filter(df.Id == 1).show()
```

```python
df.filter(df.Zipcode>10000).show()
```

```python
df.filter((df.City=="Mumbai") & (df.Branch_Number==102)).show()
```

```python
df.filter((df.City=="Mumbai") | (df.Branch_Number==100)).show()
```

```python
df.filter(df.Institution_Name.endswith("o")).show()
```

```python
df.filter(df.Branch_Name.startswith("Govind")).show()
```

```python
df.filter(df.Id.isNull()).show()
```

```python
df.filter(df.Branch_Number.isin(100,102,501)).show()
```

```python
df.filter(~df.Branch_Number.isin(100,103)).show()
```

```python
df.filter(df.Branch_Name.like("%park")).show()
```

```python
df.filter(df.Branch_Name.like("%nagar%")).show()
```

```python
df.filter(df.Branch_Name.like("Govind%")).show()
```

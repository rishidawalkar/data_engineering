```python
'''There are 3 types of Corrupt records handling:
1. Permissive: Include corrupt records in seperate column
2. Drop Malformed: Ignore corrupt records
3. Fail fast: Throws exception if corrupt found
'''
import pyspark, os, sys
from pyspark.sql import *
from pyspark import SparkConf,SparkContext
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import types

spark=SparkSession.builder.getOrCreate()

df=spark.read.format('csv').option("header",True).option("inferschema", True).load("corrupt_records.csv").show()

schema= StructType([
    StructField("Month" , StringType(),True),
    StructField("Employee_count", IntegerType(), True),
    StructField("Department", StringType(), True),
    StructField("Expense", IntegerType(), True),
    StructField("_corrupt_record", StringType(), True)
])

#Only record no. 2 and 4 should have been displayed in '_corrupt_record' column as in 2nd record we have string value in 'Expense' column which should contain integer values and in 4th record, we have value in non exsiting column 
df=spark.read.format("csv").option("mode", "PERMISSIVE").option("header",True).schema(schema).load("corrupt_records.csv").show()

#Here the corrupt records should have been dropped and the only valid records should have been displayed 
df=spark.read.format("csv").option("mode","DROPMALFORMED").option("header",True).schema(schema).load("corrupt_records.csv").show()

#This should throw an exception if corrupt record is founded
df=spark.read.format("csv").option("mode","FAILFAST").option("header",True).schema(schema).load("corrupt_records.csv").show()

```
